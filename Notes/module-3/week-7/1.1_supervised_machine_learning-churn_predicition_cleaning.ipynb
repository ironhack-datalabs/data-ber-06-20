# 1. Data Cleaning

## 1.1. Missing values

```python
df.isna().sum()
```

```python
df.replace("", np.nan).isna().sum()
```

```python
df.replace(" ", np.nan).isna().sum()
```

We have 11 occurences of whitespaces in the `TotalCharges` columns. Inspect them:

```python
df[df["TotalCharges"] == " "]
```

The occurence of 11 whitespaces in the `TotalCharges` columns is most likely due to the fact that these customers have been with the company for less than 1 month (`tenure` = 0). We could impute those values by setting it to 0.

```python
df.loc[df["TotalCharges"] == " ", "TotalCharges"] = 0
```

```python
df.replace(" ", np.nan).isna().sum()
```

```python
df.replace("-", np.nan).isna().sum()
```

## 1.2. Casting data types

```python
df.dtypes
```

Convert `TotalCharges` to `float`

```python
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

Check boolean columns

```python
df.nunique()
```

We previously assumed that all those additional services (e.g. `TechSupport`) were booleans. They actually have 3 distinct values though. 

Investigate those features.

```python
# get a list of all features with 3 unique values
should_be_booleans = list(df.nunique()[df.nunique() == 3].index)
```

```python
df["MultipleLines"].value_counts()
```

```python
df["StreamingTV"].value_counts()
```

```python
df["InternetService"].value_counts()
```

```python
df[should_be_booleans].apply(pd.value_counts)
```

We can see that `InternetService` and `Contract` are categorical (with more than 2 distinct values).

```python
cat_variables = ["InternetService", "Contract"]
df[[col for col in should_be_booleans if col not in cat_variables]].apply(pd.value_counts)
```

```python
to_bool = [col for col in should_be_booleans if col not in cat_variables]
```

```python
to_bool
```

```python
df["MultipleLines"]
```

```python
df["MultipleLines"].value_counts()
```

```python
np.where(df[to_bool] == "Yes", "Yes", "No")
```

```python
df[to_bool] = np.where(df[to_bool] == "Yes", "Yes", "No")
```

```python
df[[col for col in should_be_booleans if col not in cat_variables]].apply(pd.value_counts)
```

We removed the third value for features that are supposed to be bools. We can also convert them to numerical in the same step.

```python
df[to_bool] = np.where(df[to_bool] == "Yes", 1, 0)
```

```python
df[[col for col in should_be_booleans if col not in cat_variables]].apply(pd.value_counts)
```
