```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# Splitting our dataset into train and test set

```python
df = pd.read_pickle("data/clean_dataset.pkl")
```

```python
df.shape
```

```python
len(train_test_split(df, test_size=.2, random_state=17))
```

```python
train, test = train_test_split(df, test_size=.2, random_state=17)
```

```python
train = train.reset_index(drop=True)
test = test.reset_index(drop=True)
```
