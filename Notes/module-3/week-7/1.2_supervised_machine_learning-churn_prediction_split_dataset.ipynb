### Create alternative

```python
def split_dataset(df, test_size=.2, random_state=17, save_as_pickle=True, file_suffix="banana"):
    train, test = train_test_split(df, test_size=test_size, random_state=random_state)
    train = train.reset_index(drop=True)
    test = test.reset_index(drop=True)
    if save_as_pickle:
        train.to_pickle(f"data/train_{file_suffix}.pkl")
        test.to_pickle(f"data/test_{file_suffix}.pkl")
        
    return train, test
```

```python
split_dataset(df, random_state=20)
```
