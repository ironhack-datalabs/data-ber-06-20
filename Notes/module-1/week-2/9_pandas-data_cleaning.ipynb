### Handle missing values for columns `cylinders`

````python
df.isna().sum()
````

````python
df[df["cylinders"].isna()]
````

````python
relevant_cols = ["year", "make", "model", "trany", "drive", "fuelType", "cylinders", "displ"]
df.loc[df["cylinders"].isna(), relevant_cols]
````

Impute missing values with 0

````python
df["cylinders"].isna().sum()
````

````python
df["cylinders"].fillna(0).isna().sum()
````

````python
df["cylinders"] = df["cylinders"].fillna(0)
````

````python
df["displ"] = df["displ"].fillna(0)
````

````python
df[["cylinders", "displ"]] = df[["cylinders", "displ"]].fillna(0)
````
