````python
def drop_replicates(dframe, cols = None):
    dframe_dropped = dframe.drop_duplicates(cols)
    
    rows_dropped = len(dframe) - len(dframe_dropped)
    print(f"{rows_dropped} rows replicated and would get dropped")
    
    drop_choice = input("Do you want to inspect the rows that would get dropped? (y|n)")
    
    if drop_choice == "y":
        # identify the rows that would get dropped
        dropped_indexes = set(dframe.index) - set(dframe_dropped.index)
        dropped_rows = dframe.loc[dropped_indexes,:]
        return dropped_rows
    elif drop_choice == "n":
        drop_choice_again = input("So, do you want to drop them, then, or what? (yeee|naah)")
        if drop_choice_again == "yeee":
            return dframe_dropped
    
````

````python
drop_replicates(df, ["year", "fuelType"])
````
