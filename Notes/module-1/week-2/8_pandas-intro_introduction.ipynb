## Subsetting

````python
# access individual columns as Series
df["phoenix_loudness"]
````

````python
# access multiple columns as DataFrame
df[["phoenix_loudness", "phoenix_loudness", "phoenix_loudness", "dragon_wingspan"]]
````

## 2. rows


**Subsetting based on condition <=> Filtering**

````python
# only rows with phoenix_loudness < 0.5
df["phoenix_loudness"]
````

````python
df["phoenix_loudness"] < 0.5
# this operation returns a boolean map
````

````python
df[df["phoenix_loudness"] < 0.5]
````

````python
# subsetting based on boolean map
philter = [True, False, True, False, False, False, False, False, False, False,]
df[philter]
````

**Subsetting columns and rows at once**

````python
df[["phoenix_wingspan", "phoenix_loudness"]][philter]
````

````python
df[["phoenix_wingspan", "phoenix_loudness"]][df["dragon_wingspan"] > 0.7]
````

**using `.loc[rows,cols]`**

````python
df.loc[df["dragon_wingspan"] > 0.7, ["phoenix_wingspan", "phoenix_loudness"]]
````

````python
# all rows and all columns
df.loc[:,:]
````

````python
df.loc[:5, "phoenix_wingspan"]
````

````python
df.loc[:5, ["phoenix_wingspan"]]
````

````python
df.index = range(100,110)
````

````python
df
````

````python
df.loc[:5,:]
````

````python
# reset index
df = df.reset_index(drop=True)
````

````python
df.loc[df["phoenix_wingspan"] < .5]
````

````python
df.loc[df["dragon_wingspan"] < .5]
````

````python
df.loc[(df["phoenix_wingspan"] < .5) & (df["dragon_wingspan"] < .5), ["dragon_loudness"]]
````

```sql
SELECT
    dragon_loudness
FROM df
WHERE phoenix_wingspan < .5
    AND dragon_wingspan < .5
```
