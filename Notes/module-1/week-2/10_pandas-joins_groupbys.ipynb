## 2. Add a GROUP BY

```SQL
SELECT
    p.product_category_name,
    COUNT(*) AS items_sold
FROM order_items oi
    INNER JOIN products p
    ON oi.product_id = p.product_id
GROUP BY p.product_category_name
```

**1. using `item_category` that we defined above:**

```SQL
SELECT 
    product_category_name,
    COUNT(*) AS items_sold
FROM item_category
GROUP BY product_category_name
```

````python
# this is preferred
item_category.groupby("product_category_name").agg({"product_id": "count"})
````

````python
# this is possible
item_category[["product_category_name", "product_id"]].groupby("product_category_name").count()
````

**2. using original tables**

```SQL
SELECT
    p.product_category_name,
    COUNT(*) AS items_sold
FROM order_items oi
    INNER JOIN products p
    ON oi.product_id = p.product_id
GROUP BY p.product_category_name
```

````python
categories = (order_items
              .merge(products, on="product_id")
              .groupby("product_category_name")
              .agg({"product_id": "count"}))
````

````python
categories.index
````

````python
# reset index
categories.reset_index().rename(columns={"product_id": "items_sold"})
````

````python
(order_items
 .merge(products, on="product_id")              # `INNER JOIN products p ON oi.product_id = p.product_id`
 .groupby("product_category_name")              # `GROUP BY`
 .agg({"product_id": "count"})
 .reset_index()
 .rename(columns={"product_id": "items_sold"})) # alias `AS`
````

````python
(order_items
 .merge(products, on="product_id")
 .groupby("product_category_name")
 .agg({"product_id": "count"})
 .reset_index()
 .rename(columns={"product_id": "items_sold"}))
````

Apply two aggregate functions to one column:

````python
cat_size = price_cats.groupby("price_category", as_index=False).agg({"price": [np.size, np.average]})
cat_size
````

Weird column index. We need to resolve it first

````python
# call the .to_flat_index() method on the MultiIndex
columns = cat_size.columns.to_flat_index()
````

````python
# treat it as a list of tuples
list(columns)
````

````python
# use list comprehension to convert each tuple to a string
[first + second for first, second in columns]
````

````python
# assign new list of strings as column index
cat_size.columns = [first + second for first, second in columns]
````

````python
# rename to items_sold and avg_price
cat_size = cat_size.rename({"pricesize": "items_sold",
                            "priceaverage": "avg_price"}, axis=1)
````

````python
cat_size
````

#### How could we avoid dealing with MultiIndex?

````python
# 1. duplicate the price column with a different name
cat_aw = price_cats.assign(something_else=price_cats["price"])
cat_aw
````

````python
# 2. apply separate aggregations to those two columns
(cat_aw
 .groupby("price_category", as_index=False)
 .agg({"price": np.size,
       "something_else": np.average})
 .rename(columns={"price": "items_sold",
                  "something_else": "avg_price"}))
````
